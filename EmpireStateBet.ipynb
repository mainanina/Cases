```python
import numpy as np
```

```python
all_walks = []
```

```python
for i in range(250):
  # starting random walk
  random_walk = [0]
  for _ in range(50):
    # setting the previous step 
    step = random_walk[-1]
    # rolling the dice
    dice = np.random.randint(1,7)
    # bet rules
    if dice <= 2:
      step = max(0, step - 1)
    elif dice <=5:
      step = step +1
    else: 
      step = step + np.random.randint(1,7)
    # adding the chance of falling (and start from the begining)
    if np.random.rand()<=0.001:
      step = 0
    random_walk.append(step)
  all_walks.append(random_walk)
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting all the random_walks
np_aw = np.array(all_walks)
np_aw_t = np.transpose(np_aw)
plt.plot(np_aw_t)
plt.show()
```

```python
# storing the last step from every random walk
ends = np_aw_t[-1, :]
# distribuition of last steps
plt.hist(ends)
plt.show()
```

```python
# calculating the estimated chance of reaching 50 steps
round((len(ends[ends>=50])/len(ends))*100, 2)
```
